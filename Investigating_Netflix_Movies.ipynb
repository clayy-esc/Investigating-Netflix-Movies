```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# import Netflix data as DataFrame
netflix_df = pd.read_csv("netflix_data.csv", index_col=0)

netflix_movies = netflix_df[netflix_df['type'] == 'Movie']
movies_by_decade = netflix_movies[np.logical_and(netflix_movies['release_year'] >= 2000, netflix_movies['release_year'] <= 2010)].copy()

def get_label_duration(duration):
    if duration < 90:
        return 'Short'
    elif duration < 180:
        return 'Medium'
    else:
        return 'Long'

movies_by_decade['label_duration'] = movies_by_decade['duration'].apply(get_label_duration)
```

```python
# Count No. of Movies by Genre
plt.figure(figsize=(12, 6))

genre_counts = movies_by_decade['genre'].value_counts()
plt.bar(genre_counts.index, genre_counts.values, color='skyblue', edgecolor='black')

plt.title('Number of Movies by Genre (2000 - 2010)', fontsize=18, fontweight='bold', color='skyblue')
plt.xlabel('Genre', fontsize=14, fontweight='bold')
plt.ylabel('Number of Movies',fontsize=14, fontweight='bold')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', linewidth=0.5, alpha=0.7)
plt.show()

print('Number of Movies by Genre (2000 - 2010)')
for lab, row in genre_counts.items():
    print(f'{lab}: {row}')

plt.clf()
```

```python
# Count No. of Movies by Duration
plt.figure(figsize=(12, 6))

duration_counts = movies_by_decade['label_duration'].value_counts()
plt.bar(duration_counts.index, duration_counts.values, color=['lightgreen', 'orange', 'salmon'], edgecolor='black')

plt.title('Number of Movies by Duration (2000 - 2010)', fontsize=18, fontweight='bold')
plt.xlabel('Duration', fontsize=14, fontweight='bold')
plt.ylabel('Number of Movies', fontsize=14, fontweight='bold')
plt.grid(axis='y', linestyle='--', linewidth=0.5, alpha=0.7)
plt.show()

print('Number of Movies by Duration (2000 - 2010)')
for lab, row in duration_counts.items():
    print(f'{lab}: {row}')

plt.clf()
```

```python
# Get the most preferred Movie Genre with the most preferred Duration
preferred_movie = movies_by_decade[movies_by_decade['genre'] == genre_counts.index[0]]

preferred_movie_count = 0
for lab, row in preferred_movie.iterrows():
    if row['label_duration'] == duration_counts.index[0]:
        preferred_movie_count += 1

print('Most Preferred Movie to Watch from 2000 to 2010')
print(f'Genre: {genre_counts.index[0]}')
print(f'Duration: {duration_counts.index[0]}')
print(f'Number of Movies: {preferred_movie_count}')
```
